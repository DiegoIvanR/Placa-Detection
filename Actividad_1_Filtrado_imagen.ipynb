```python
import cv2
import numpy as np

from google.colab import files
import matplotlib.pyplot as plt
```

```python
def color_filter(img, b):
  for i in range(img.shape[0]):  # filtro q me invente para no perder tanto informacion del negro la neta no se como funciona pero funciona dos tres
    for j in range(img.shape[1]):
      if img[i, j] > b:
        img[i, j] = 255
      else:
        img[i, j] = np.interp(img[i, j], [0, b], [0, 200])
  return img
```

```python
def black_filter(img, b):
  for i in range(img.shape[0]):  # filtro negro
    for j in range(img.shape[1]):
      if img[i, j] > b:
        img[i, j] = 255
      else:
        img[i, j] = 0
  return img
```

```python
img_path = '/content/placa_4.jpg'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
rgb_filter = color_filter(img, 125)
plt.imshow(rgb_filter, cmap='gray')
plt.show()
```

```python
img_blur3 = cv2.blur(src=rgb_filter, ksize=(15, 15))
filter2 = black_filter(img_blur3, 200)
plt.imshow(filter2, 'gray')
plt.show()
```

```python
img_blur4 = cv2.blur(src=filter2, ksize=(7, 7))

filter3 = black_filter(img_blur4, 20)
plt.imshow(filter3, 'gray')
plt.show()
```

```python
img_blur5 = cv2.blur(src=filter3, ksize=(7, 7))

filter4 = black_filter(img_blur5, 10)
plt.imshow(filter4, 'gray')
plt.show()
```

```python
img_blur6 = cv2.blur(src=filter4, ksize=(7, 7))

filter5 = black_filter(img_blur6, 10)
plt.imshow(filter5, 'gray')
plt.show()
```

```python
canny = cv2.Canny(filter5, threshold1=40, threshold2=100)
plt.imshow(canny, 'gray')
plt.show()
```
